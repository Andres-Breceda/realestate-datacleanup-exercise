#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import matplotlib.pyplot as plt

# Asegúrate de tener el subconjunto del Ejercicio 14
subconjunto = ds[ds['level5'].isin(['Getafe', 'Fuenlabrada', 'Leganés', 'Alcorcón'])]

# Normalización de los precios para cada población con .loc[]
subconjunto.loc[:, 'price_normalized'] = subconjunto.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Verifica los datos de normalización
print(subconjunto[['level5', 'price', 'price_normalized']].head())
print(subconjunto['price_normalized'].describe())  # Para ver el rango de la normalización

# Crear subgráficos para cada población
fig, axes = plt.subplots(1, 4, figsize=(20, 5))  # 1 fila, 4 columnas

# Trazar un histograma para cada población en su subgráfico
for i, poblacion in enumerate(['Getafe', 'Fuenlabrada', 'Leganés', 'Alcorcón']):
    data = subconjunto[subconjunto['level5'] == poblacion]['price_normalized']
    if len(data) > 0:  # Verificar que haya datos para esta población
        axes[i].hist(data, bins=20, alpha=0.7)
        axes[i].set_title(f'Población {poblacion}')
        axes[i].set_xlabel('Precio Normalizado')
        axes[i].set_ylabel('Frecuencia')
    else:
        axes[i].set_title(f'Población {poblacion} (Sin datos)')
        axes[i].set_xlabel('Precio Normalizado')
        axes[i].set_ylabel('Frecuencia')

# Ajustar el espacio entre los subgráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# Filtrar las poblaciones de Getafe y Alcorcón
subconjunto_getafe_alcorcon = subconjunto[subconjunto['level5'].isin(['Getafe', 'Alcorcón'])]

# Calcular el precio por metro cuadrado
subconjunto_getafe_alcorcon['price_per_m2'] = subconjunto_getafe_alcorcon['price'] / subconjunto_getafe_alcorcon['surface']

# Mostrar los primeros registros de los precios por metro cuadrado
print(subconjunto_getafe_alcorcon[['level5', 'price', 'surface', 'price_per_m2']].head())

# Comparar el precio medio por metro cuadrado entre Getafe y Alcorcón
precio_medio_getafe = subconjunto_getafe_alcorcon[subconjunto_getafe_alcorcon['level5'] == 'Getafe']['price_per_m2'].mean()
precio_medio_alcorcon = subconjunto_getafe_alcorcon[subconjunto_getafe_alcorcon['level5'] == 'Alcorcón']['price_per_m2'].mean()

print(f"Precio medio por metro cuadrado en Getafe: {precio_medio_getafe}")
print(f"Precio medio por metro cuadrado en Alcorcón: {precio_medio_alcorcon}")
print("El precio de los medio de los metros cuadrados en Alcorcon es mas caro que en Getafe")
```

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python

import matplotlib.pyplot as plt

# Filtrar las 4 poblaciones (seleccionadas por ejemplo: 'Getafe', 'Fuenlabrada', 'Leganés', 'Alcorcón')
poblaciones = ['Getafe', 'Fuenlabrada', 'Leganés', 'Alcorcón']
subconjunto_poblaciones = subconjunto[subconjunto['level5'].isin(poblaciones)]

# Crear los subgráficos (2 filas, 2 columnas para 4 poblaciones)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Iterar sobre las poblaciones y crear un gráfico de dispersión para cada una
for i, poblacion in enumerate(poblaciones):
    ax = axes[i//2, i%2]  # Elegir la posición en el subplot (2x2)
    df_poblacion = subconjunto_poblaciones[subconjunto_poblaciones['level5'] == poblacion]
    
    # Crear el gráfico de dispersión para cada población
    ax.scatter(df_poblacion['price'], df_poblacion['surface'], label=f'{poblacion}')
    ax.set_title(f'Diagrama de Dispersión - {poblacion}')
    ax.set_xlabel('Precio')
    ax.set_ylabel('Superficie')
    ax.legend()

# Ajustar el diseño del gráfico
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps, Marker, LayerGroup
from ipywidgets import HTML

# Datos actualizados del cinturón sur de Madrid con coordenadas precisas
data = {
    'latitude': [40.30571, 40.32718, 40.28419, 40.34582, 40.23604],
    'longitude': [-3.73295, -3.7635, -3.79415, -3.82487, -3.76752],
    'poblacion': ['Getafe', 'Leganés', 'Fuenlabrada', 'Alcorcón', 'Parla']
}

# Definir colores para cada población
colores = {
    'Getafe': 'red',
    'Leganés': 'blue',
    'Fuenlabrada': 'green',
    'Alcorcón': 'orange',
    'Parla': 'purple'
}

# Crear el mapa centrado en el cinturón sur de Madrid
mapa = Map(center=(40.32, -3.78), zoom=11, min_zoom=1, max_zoom=20, basemap=basemaps.OpenTopoMap)

# Agregar marcadores con colores según la población
marcadores = []
for lat, lon, poblacion in zip(data['latitude'], data['longitude'], data['poblacion']):
    marker = Marker(location=(lat, lon), draggable=False)
    popup = HTML()
    popup.value = f"<b>{poblacion}</b>"
    marker.popup = popup
    marker.icon_color = colores[poblacion]
    marcadores.append(marker)

# Agregar los marcadores al mapa
grupo = LayerGroup(layers=marcadores)
mapa.add_layer(grupo)

# Mostrar el mapa
mapa
```
